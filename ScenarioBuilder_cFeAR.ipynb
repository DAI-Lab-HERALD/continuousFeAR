# Check FeAR

```python
# Hyper Parameters for faster runs

l = 2
T_ACTION=5
# N_TIMESTEPS=5
N_TIMESTEPS=10 

A_NUM = 10
THETA_NUM = 32

# A_NUM = 4
# THETA_NUM = 8

A_MIN = 0
A_MAX = 5

THETA_MIN = -np.pi
THETA_MAX = np.pi


t_action = T_ACTION
n_timesteps = N_TIMESTEPS

a_num = A_NUM
theta_num = THETA_NUM

a_min = A_MIN
a_max = A_MAX 

theta_min = THETA_MIN
theta_max = THETA_MAX

plot_scenario = True
collision_free_trajs = False
```

```python
#  Creating the trajectories for the initial conditions

x_, y_, t_ = cfear.make_trajectories(x0=x0, y0=y0, v0x=v0x, v0y=v0y, a=a, theta=theta,
                               t_action=t_action, n_timesteps=n_timesteps)

trajs_boxes = cfear.get_boxes_for_trajs(x_, y_, lx=l, ly=l)


trajs_hulls = cfear.get_trajs_hulls(trajs_boxes)

if plot_scenario:
    ax = plotcf.plot_hulls_for_trajs(trajs_hulls, trajs_boxes=trajs_boxes, colour='lightgrey')
    plotcf.plot_obstacle(obstacle, ax=ax)

    ax = plotcf.plot_velocities_for_agents(x=x0, y=y0, vx=v0x, vy=v0y, ax=ax, scale=t_action / n_timesteps)

# -------------------------------------------------------------------------------------------------------------

# Check for collisions and rectify collisions

trajs_hulls_no_collision, trajs_boxes_no_collision = cfear.get_collision_free_trajs(trajs_hulls, trajs_boxes, obstacle=obstacle,
                                                    plot_polygons_one_by_one=False, verbose_flag=True)

ax = plotcf.plot_boxes_for_trajs(trajs_boxes_no_collision, obstacle=obstacle)
ax = plotcf.plot_hulls_for_trajs(trajs_hulls_no_collision, obstacle=obstacle)

# -------------------------------------------------------------------------------------------------------------
```

```python
# Compute FeAR

fear = cfear.get_fear(nn=N, mdr_a=mdr_a, mdr_theta=mdr_theta,
                a_min=a_min, a_max=a_max, a_num=a_num,
                theta_min=theta_min, theta_max=theta_max, theta_num=theta_num,
                trajs_hulls=trajs_hulls,
                trajs_boxes=trajs_boxes,
                obstacle=obstacle,
                x0=x0, y0=y0, v0x=v0x, v0y=v0y,
                t_action=t_action, n_timesteps=n_timesteps,
                l=l, plot_subspaces=False,
                plot_feasible_action_space=False,
                return_feasibility=False)
```

```python
print(f'{fear=}')

if plot_scenario:
    ax = plotcf.plot_hulls_for_trajs(trajs_hulls, trajs_boxes=trajs_boxes, colour='lightgrey')
    plotcf.plot_obstacle(obstacle, ax=ax)

    ax = plotcf.plot_velocities_for_agents(x=x0, y=y0, vx=v0x, vy=v0y, ax=ax, scale=t_action / n_timesteps)

plotcf.plot_fear(fear, fmt= '.1f')
```
